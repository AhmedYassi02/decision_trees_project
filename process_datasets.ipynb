```python
import pandas as pd
```

```python
prima_indians = pd.read_csv("data/prima_indians.csv")
thyroid_diff = pd.read_csv("data\Thyroid_Diff.csv")
tcga_mutations = pd.read_csv("data\TCGA_GBM_LGG_Mutations_all.csv")
```

```python
prima_indians_X = prima_indians.drop(columns=["Outcome"])
prima_indians_y = prima_indians["Outcome"]
```

```python
# stor in txt format as : 
# X = [ row 1, row 2, ...]
# Y =  prima_indians_y

with open("data/prima_indians.txt", "w") as f:
    f.write("X = [")
    for index, row in prima_indians_X.iterrows():
        f.write(" ".join(map(str, row.values)))
        f.write(";")
    f.write("]\n")

    f.write("Y = [\n")
    for index, value in enumerate(prima_indians_y):
        f.write(f"  {value},")
    f.write("]\n")
```

```python

# dummy variables for all variables of thyroid_diff
thyroid_diff_dummies = pd.get_dummies(thyroid_diff, drop_first=True)*1
# thyroid_diff
```

```python
thyroid_diff_X  = thyroid_diff_dummies.drop(columns=["Recurred_Yes"])
thyroid_diff_y = thyroid_diff_dummies["Recurred_Yes"]
```

```python
with open("data/thyroid_diff.txt", "w") as f:
    f.write("X = [")
    for index, row in thyroid_diff_X.iterrows():
        f.write(" ".join(map(str, row.values)))
        f.write(";")
    f.write("]\n")

    f.write("Y = [")
    for index, value in enumerate(thyroid_diff_y):
        f.write(f"  {value},")
    f.write("]")
```
